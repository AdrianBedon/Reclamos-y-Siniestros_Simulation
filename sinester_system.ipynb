```python
import pandas as pd
import numpy as np
import os
import time
import datetime
```

```python
folder_path = r'\\192.168.3.252\reclamos_siniestros'
file = "Reclamos.csv"
```

```python
def monitor_folder(folder_path):
    file_mod_time = os.path.getmtime(os.path.join(folder_path,file))
    print(f"Last modification for {file}: {datetime.datetime.fromtimestamp(file_mod_time)}")

    while True:
        time.sleep(30)
        if (os.path.exists(os.path.join(folder_path,file))):
            current_file_mod_time = os.path.getmtime(os.path.join(folder_path,file))
            if current_file_mod_time != file_mod_time:
                siniestros_df = pd.read_csv(os.path.join(folder_path,file))
                display(siniestros_df)
                siniestros_df['Status'] = np.random.choice(['aprobed', 'denied'], size=len(siniestros_df))
                siniestros_df.to_csv(os.path.join(folder_path,"Reclamos_Siniestros_History.csv"), mode='a', header=False, index=False)
                siniestros_df.to_csv(os.path.join(folder_path,"Siniestros.csv"), index=False)
                os.remove(os.path.join(folder_path,file))
                file_mod_time = current_file_mod_time
        print(f"Last modification for {file}: {datetime.datetime.fromtimestamp(file_mod_time)}")
```

```python
monitor_folder(folder_path)
```
