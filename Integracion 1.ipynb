```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd
from openpyxl import Workbook, load_workbook
import os
```

```python
# Crear ventana principal
root = tk.Tk()
root.title("Formulario de Reclamos")

# Función para guardar los datos en un archivo Excel
def guardar_datos():
    num_poliza = entry_num_poliza.get()
    fecha_reclamo = entry_fecha_reclamo.get()
    descripcion = entry_descripcion.get("1.0", tk.END)
    monto = entry_monto.get()
    
    # Verificar si todos los campos están completos
    if not (num_poliza and fecha_reclamo and descripcion and monto):
        messagebox.showerror("Error", "Todos los campos deben estar completos")
        return
    
    # Definir ruta de carpeta compartida
    file_path = r'\\192.168.3.252\reclamos_siniestros'
    if not os.path.exists(file_path):
        os.makedirs(file_path)
        
    archivo_excel = os.path.join(file_path, r'\Reclamos.xlsx')
    
    # Crear o cargar archivo Excel
    if not os.path.exists(archivo_excel):
        wb = Workbook()
        ws = wb.active
        ws.append(["Número de Póliza", "Fecha del Reclamo", "Descripción del Incidente", "Monto del Reclamo"])
        wb.save(archivo_excel)
    
    # Agregar datos al archivo Excel
    wb = load_workbook(archivo_excel)
    ws = wb.active
    ws.append([num_poliza, fecha_reclamo, descripcion.strip(), monto])
    wb.save(archivo_excel)

    reclamos_df = pd.read_excel(archivo_excel)
    reclamos_df.to_csv(r'\\192.168.3.252\reclamos_siniestros\prueba.csv')
    
    # Limpiar campos del formulario
    entry_num_poliza.delete(0, tk.END)
    entry_fecha_reclamo.delete(0, tk.END)
    entry_descripcion.delete("1.0", tk.END)
    entry_monto.delete(0, tk.END)
    
    messagebox.showinfo("Éxito", "Datos guardados correctamente")
```

```python
# Etiquetas y campos de entrada
tk.Label(root, text="Número de Póliza:").grid(row=0, column=0, padx=10, pady=5)
entry_num_poliza = tk.Entry(root)
entry_num_poliza.grid(row=0, column=1, padx=10, pady=5)

tk.Label(root, text="Fecha del Reclamo:").grid(row=1, column=0, padx=10, pady=5)
entry_fecha_reclamo = tk.Entry(root)
entry_fecha_reclamo.grid(row=1, column=1, padx=10, pady=5)

tk.Label(root, text="Descripción del Incidente:").grid(row=2, column=0, padx=10, pady=5)
entry_descripcion = tk.Text(root, height=5, width=40)
entry_descripcion.grid(row=2, column=1, padx=10, pady=5)

tk.Label(root, text="Monto del Reclamo:").grid(row=3, column=0, padx=10, pady=5)
entry_monto = tk.Entry(root)
entry_monto.grid(row=3, column=1, padx=10, pady=5)

# Botones
tk.Button(root, text="Guardar Datos", command=guardar_datos).grid(row=4, column=0, columnspan=2, pady=10)
#tk.Button(root, text="Leer Excel", command=leer_excel).grid(row=5, column=0, columnspan=2, pady=10)
root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox
import csv
import os
from tkinter import ttk
```

```python
# Función para guardar los datos en un archivo CSV
def guardar_datos():
    num_poliza = entry_num_poliza.get()
    fecha_reclamo = entry_fecha_reclamo.get()
    descripcion = entry_descripcion.get("1.0", tk.END)
    monto = entry_monto.get()
    
    # Verificar si todos los campos están completos
    if not (num_poliza and fecha_reclamo and descripcion and monto):
        messagebox.showerror("Error", "Todos los campos deben estar completos")
        return
    
    # Definir ruta de carpeta compartida
    ruta_carpeta = r'\\192.168.3.252\reclamos_siniestros'
    #ruta_carpeta = 'C:/Ratios/'
    if not os.path.exists(ruta_carpeta):
        os.makedirs(ruta_carpeta)
        
    archivo_csv = os.path.join(ruta_carpeta, 'Reclamos.csv')

    # Verificar si el archivo CSV ya existe
    file_exists = os.path.isfile(archivo_csv)
    
    # Abrir el archivo CSV y escribir los datos
    with open(archivo_csv, mode='a', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        
        # Si el archivo no existía, escribir el encabezado
        if not file_exists:
            writer.writerow(["Número de Póliza", "Fecha del Reclamo", "Descripción del Incidente", "Monto del Reclamo"])
        
        # Escribir los datos ingresados
        writer.writerow([num_poliza, fecha_reclamo, descripcion.strip(), monto])
    
    # Limpiar campos del formulario
    entry_num_poliza.delete(0, tk.END)
    entry_fecha_reclamo.delete(0, tk.END)
    entry_descripcion.delete("1.0", tk.END)
    entry_monto.delete(0, tk.END)
    
    messagebox.showinfo("Éxito", "Datos guardados correctamente")
```

```python
def leer_csv():
    archivo_a_leer = r'\\192.168.3.252\reclamos_siniestros/Siniestros.csv'

    if not os.path.exists(archivo_a_leer):
        messagebox.showerror("Error", "El archivo especificado no existe")
        return

    # Create a new window to display the table
    table_window = tk.Tk()
    table_window.title("Datos del Archivo")

    # Set up style to change the header color
    style = ttk.Style()
    style.theme_use("default")
    style.configure("Treeview.Heading", background="gray", foreground="black", font=('Arial', 10, 'bold'))

    # Create a Treeview widget to show the table data
    tree = ttk.Treeview(table_window)
    
    # Read the CSV file
    with open(archivo_a_leer, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        datos = list(reader)

        # Set up columns based on the CSV header (assuming the first row contains headers)
        if len(datos) > 0:
            tree["columns"] = datos[0]
            tree["show"] = "headings"

            # Define columns and set their headings
            for col in datos[0]:
                tree.heading(col, text=col)
                tree.column(col, anchor='center')

            # Add data rows to the Treeview
            for fila in datos[1:]:
                tree.insert("", "end", values=fila)

    # Pack the Treeview widget into the window
    tree.pack(expand=True, fill="both")

    # Start the GUI loop
    table_window.mainloop()
```

```python
def leer_csvHistorico():
    archivo_a_leer = r'\\192.168.3.252\reclamos_siniestros/Reclamos_Siniestros_History.csv'

    if not os.path.exists(archivo_a_leer):
        messagebox.showerror("Error", "El archivo especificado no existe")
        return

    table_window = tk.Tk()
    table_window.title("Datos del Archivo")

    tree = ttk.Treeview(table_window)
    
    with open(archivo_a_leer, mode='r', newline='', encoding='utf-8') as file:
        reader = csv.reader(file)
        datos = list(reader)

        if len(datos) > 0:
            tree["columns"] = datos[0]
            tree["show"] = "headings"

            for col in datos[0]:
                tree.heading(col, text=col)
                tree.column(col, anchor='center')

            for fila in datos[1:]:
                tree.insert("", "end", values=fila)

    tree.pack(expand=True, fill="both")

    table_window.mainloop()
```

```python
# Crear ventana principal
root = tk.Tk()
root.title("Formulario de Reclamos")

# Etiquetas y campos de entrada
tk.Label(root, text="Número de Póliza:").grid(row=0, column=0, padx=10, pady=5)
entry_num_poliza = tk.Entry(root)
entry_num_poliza.grid(row=0, column=1, padx=10, pady=5)

tk.Label(root, text="Fecha del Reclamo:").grid(row=1, column=0, padx=10, pady=5)
entry_fecha_reclamo = tk.Entry(root)
entry_fecha_reclamo.grid(row=1, column=1, padx=10, pady=5)

tk.Label(root, text="Descripción del Incidente:").grid(row=2, column=0, padx=10, pady=5)
entry_descripcion = tk.Text(root, height=5, width=40)
entry_descripcion.grid(row=2, column=1, padx=10, pady=5)

tk.Label(root, text="Monto del Reclamo:").grid(row=3, column=0, padx=10, pady=5)
entry_monto = tk.Entry(root)
entry_monto.grid(row=3, column=1, padx=10, pady=5)

# Botones
tk.Button(root, text="Guardar Reclamo", command=guardar_datos).grid(row=4, column=0, columnspan=2, pady=10)
tk.Button(root, text="Leer reclamos procesados", command=leer_csv).grid(row=5, column=0, columnspan=2, pady=10)
tk.Button(root, text="Leer reclamos procesados históricos", command=leer_csvHistorico).grid(row=6, column=0, columnspan=2, pady=10)

# Ejecutar la ventana principal
root.mainloop()
```
